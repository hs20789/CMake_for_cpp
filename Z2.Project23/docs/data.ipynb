```python
import matplotlib.pyplot as plt
import csv
```

```python
# CSV 파일 읽기
x_vals = []
y_vals = []

with open("res1.csv", encoding="utf-8") as f:
    reader = csv.reader(f)
    for row in reader:
        # row는 ["1", "10130"] 이런 형태의 문자열 리스트
        x_vals.append(int(row[0]))
        y_vals.append(int(row[1]))
        
```

```python
# 그래프 그리기
plt.plot(x_vals, y_vals, marker='o', linestyle='-', color='blue', label="Data")
plt.xlabel("Index")
plt.ylabel("Value")
plt.ylim(0, 12000)
plt.title("CSV Data Plot")
plt.legend()
plt.grid(True)
plt.show()
```
