```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = pd.read_csv("res1.csv", encoding='utf-8', names=['x', 'y'])

plt.plot(df1['x'], df1['y'])
plt.title('Normal Distribution')
plt.show()
```

```python
df2 = pd.read_csv("res2.csv", encoding='utf-8', names=['x', 'y'])

plt.plot(df2['x'], df2['y'], c='hotpink')
plt.title("Exponential Distribution")
plt.show()
```
